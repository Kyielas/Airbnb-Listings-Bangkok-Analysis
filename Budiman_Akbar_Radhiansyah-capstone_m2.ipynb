# **Capstone Project : <u>Airbnb Listings Bangkok</u>**

## **Background**
Airbnb, a global online platform facilitating short-term lodging rentals, encompasses diverse accommodation options in numerous cities worldwide, including Bangkok, Thailand. The dataset in focus pertains specifically to Airbnb listings in Bangkok, providing comprehensive details about various accommodations, guest reviews, geographical distribution, and attributes associated with these listings.

## **Problem Statement**
The dataset's objective is to analyze Airbnb Listings in Bangkok to extract insights for improving policies, enhancing host-guest trust, and optimizing the Airbnb experience. By identifying trends, understanding guest preferences, and addressing issues such as inadequate reviews or negative feedback, this analysis aims to guide Airbnb Bangkok in devising better strategies for hosts, renters, and guests. The fundamental question this analysis seeks to answer is:

**"How Do Varied Booking Preferences Unfold Among Travelers on a Airbnb Rental Platform?"**

Understanding user preferences and booking trends is crucial for improving customer experience and optimizing the platform's offerings. Both of these distinctions can assist in tailoring services, refining marketing strategies, and fostering a more conducive environment for both hosts and guests within the Airbnb ecosystem in Bangkok.

## **Booking Frequencies Over Time**

```python
df_copy = df.copy()  # Create a copy of the original DataFrame

# Convert 'last_review' column to datetime format in the copied DataFrame
df_copy['last_review'] = pd.to_datetime(df_copy['last_review'])

# Set 'last_review' as the index
df_copy.set_index('last_review', inplace=True)

# Filter out records with 'last_review' after 1910
df_filtered = df_copy[df_copy.index.year >= 1910]

# Resample monthly and plot the bookings over time
bookings_over_time = df_filtered.resample('M').size()  # Resample monthly
bookings_over_time.plot(figsize=(12, 6),color='red', linestyle='-')
plt.title('Booking Frequencies Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Number of Bookings', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_copy2 = df.copy()  # Create a copy of the original DataFrame

# Convert 'last_review' column to datetime format in the copied DataFrame
df_copy2['last_review'] = pd.to_datetime(df_copy2['last_review'], errors='coerce')

# Filter out records with 'last_review' between 2013 and 2022 in the copied DataFrame
df_filtered2 = df_copy2[(df_copy2['last_review'].dt.year >= 2013) & (df_copy2['last_review'].dt.year <= 2022)]

# Create subplots for each year
years = range(2013, 2023)
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8))
for i, year in enumerate(years):
    row = i // 5
    col = i % 5
    ax = axes[row, col]

    bookings_over_time = df_filtered2[df_filtered2['last_review'].dt.year == year].resample('M', on='last_review').size()  # Resample monthly
    bookings_over_time.plot(ax=ax, title=f'Booking Frequencies - {year}', legend=False)
    ax.set_xlabel('Date')
    ax.set_ylabel('Number of Bookings')
    ax.grid(True)

plt.tight_layout()
plt.show()
```

### **Insights**

**1. Growth in Bookings Over the Years:**

There is a visible growth trend in the number of bookings over the years, with a gradual increase starting from 2013 to 2022.
Particularly notable is the significant surge in bookings from 2018 onwards, indicating an increased popularity or higher demand for accommodations during this period.

**2. Peak Booking Months:**

December of each year consistently appears to be a peak month for bookings, suggesting high demand, possibly due to holiday seasons, festivals, or year-end vacations.
The months of January and February also exhibit higher booking numbers in later years, possibly due to post-holiday travel or winter tourism.

**3. Impact of External Factors:**

An unexpected drop in bookings is observed in April 2020, indicating a potential impact of external factors such as the COVID-19 pandemic, which led to reduced travel and accommodation bookings globally during that period.

**4. Accelerated Booking Growth in Recent Years:**

The highest number of bookings is observed in December 2022, indicating a substantial surge in demand, potentially attributed to various factors like eased travel restrictions, increased travel confidence, or special events occurring during that time.

```python
# Create an empty list to store the number of bookings per month
booking_data = []

# Iterate through each year and month to count bookings
for year in range(2013, 2023):
    for month in range(1, 13):  # Iterate through months (1 to 12)
        bookings_per_month = df_filtered2[
            (df_filtered2['last_review'].dt.year == year) & 
            (df_filtered2['last_review'].dt.month == month)
        ].shape[0]  # Count bookings for the specific year and month
        
        # Append data to the list
        booking_data.append({
            'Year': year,
            'Month': month,
            'Number_of_Bookings': bookings_per_month
        })

# Create a DataFrame from the collected data
booking_counts = pd.DataFrame(booking_data)

# Display the DataFrame containing the number of bookings per month
display(booking_counts[booking_counts['Year']==2020],booking_counts[booking_counts['Year']==2021],booking_counts[booking_counts['Year']==2022])
```

## **Room Type Preferences**

```python
years = range(2013, 2023)
num_rows = 5
num_cols = 2

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12, 18))

for i, year in enumerate(years):
    row = i // num_cols
    col = i % num_cols
    
    # Filter data for the current year
    df_year = df[df['last_review'].dt.year == year]
    
    # Calculate room type counts for the current year
    room_type_counts = df_year['room_type'].value_counts()
    
    ax = axes[row, col]
    sns.barplot(x=room_type_counts.index, y=room_type_counts.values, hue=room_type_counts.index, ax=ax)
    ax.set_title(f'Room Type Popularity in {year}')
    ax.set_xlabel('Room Type')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', labelrotation=45)  # Rotate x-labels for better readability

plt.tight_layout()
plt.show()
```

### **Insight**

**1. Shift in Room Type Preferences:**

There's a consistent trend of increased popularity in "Entire home/apt" over the years, indicating a growing preference for more private accommodations.
The preference for "Private room" shows relative stability but experiences fluctuations, suggesting moderate user interest in this room type.

**2. Emergence of "Shared room" Preference:**

Although less popular compared to other room types, there's a recurring count for "Shared room" bookings across different years, indicating a consistent niche market or user preference for more economical lodging options.

**3. Long-Term Patterns in Room Type Preferences:**

Long-term analysis showcases a shift in preferences, such as the decline of "Shared room" bookings over the years and the rise of "Entire home/apt" preferences, reflecting changes in consumer behavior or market dynamics.

```python
years = range(2013, 2023)
room_type_counts_data = []

for year in years:
    for month in range(1, 13):
        # Filter data for the current year and month
        df_month = df[(df['last_review'].dt.year == year) & (df['last_review'].dt.month == month)]
        
        # Calculate room type counts for the current year and month
        room_type_counts = df_month['room_type'].value_counts().to_dict()
        
        # Append the counts to the list
        room_type_counts_data.append({
            'Year': year,
            'Month': month,
            **room_type_counts
        })

# Create a DataFrame from the collected data
room_type_counts_df = pd.DataFrame(room_type_counts_data)

room_type_counts_df
display(room_type_counts_df[room_type_counts_df['Year']==2022],room_type_counts_df[room_type_counts_df['Year']==2020],room_type_counts_df[room_type_counts_df['Year']==2021])
```

## **Location Preferences**

```python
from folium.plugins import HeatMap

# Calculate booking frequency per neighborhood
neighborhood_booking_freq = df['neighbourhood'].value_counts()

# Create a map centered around the mean latitude and longitude
mean_lat = df['latitude'].mean()
mean_lon = df['longitude'].mean()
m = folium.Map(location=[mean_lat, mean_lon], zoom_start=12)

# Get the mean latitude and longitude per neighborhood for heatmap
heatmap_data = []
for neighborhood, count in neighborhood_booking_freq.items():
    subset = df[df['neighbourhood'] == neighborhood]
    neighborhood_lat = subset['latitude'].mean()
    neighborhood_lon = subset['longitude'].mean()
    heatmap_data.append([neighborhood_lat, neighborhood_lon, count])

# Create a HeatMap with booking frequencies per neighborhood
HeatMap(heatmap_data).add_to(m)

# Display the map
m
```

```python
# Calculate booking frequency per neighborhood
neighborhood_booking_freq = df['neighbourhood'].value_counts()

# Get the mean latitude and longitude per neighborhood for heatmap
heatmap_data = []
for neighborhood, count in neighborhood_booking_freq.items():
    subset = df[df['neighbourhood'] == neighborhood]
    neighborhood_lat = subset['latitude'].mean()
    neighborhood_lon = subset['longitude'].mean()
    heatmap_data.append([neighborhood, neighborhood_lat, neighborhood_lon, count])  # Include neighborhood name

# Create a DataFrame based on the heatmap_data
columns = ['Neighborhood', 'Neighborhood Latitude', 'Neighborhood Longitude', 'Booking Frequency']
heatmap_df = pd.DataFrame(heatmap_data, columns=columns)

# Display the DataFrame
heatmap_df
```

```python
# Sort the DataFrame by 'Booking Frequency' in descending order
sorted_heatmap_df = heatmap_df.sort_values(by='Booking Frequency', ascending=False)

# Get the top 5 neighborhoods by booking frequency
top_5_neighborhoods = sorted_heatmap_df.head(5)

# Create a bar plot for top 5 neighborhoods by booking frequency with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Neighborhood', y='Booking Frequency', hue='Neighborhood', data=top_5_neighborhoods, palette='viridis', errorbar=('ci',0), dodge=False)
plt.xlabel('Neighborhood')
plt.ylabel('Booking Frequency')
plt.title('Top 5 Neighborhoods by Booking Frequency')
plt.legend([], frameon=False)  # Hide legend

# Annotate the bars with their values
for index, row in top_5_neighborhoods.iterrows():
    plt.text(row.name, row['Booking Frequency'] + 1, str(row['Booking Frequency']), ha='center', va='bottom', fontsize=10)

plt.xticks(rotation=45, ha='right')  # Rotate x-labels for better readability
plt.tight_layout()
plt.show()
```

**Booking Patterns and Trends:**

* A consistent increase in bookings has been witnessed over the years, notably peaking in December and showing signs of recovery after a drop in 2020 due to the global pandemic.
* The preference for entire home/apartment accommodations has steadily increased, reflecting a shift towards more private lodging options.

**Location Preferences and Popularity:**

* Neighborhoods like Vadhana, Khlong Toei, and Bang Rak stand out with higher booking frequencies, suggesting their popularity among travelers. These areas might serve as potential tourist hotspots due to various attractions or convenient amenities.
* Geographical distribution shows varied booking frequencies across different parts of Bangkok, highlighting distinct traveler interests or accommodation availability.

**Price Sensitivity and Neighborhood Variation:**

* Price sensitivity across neighborhoods is evident, with lower-priced areas like Bang Bon attracting budget-conscious travelers, while affluent districts like Khlong San cater to luxury-seeking clientele.
* Varied property types, amenities, and proximity to attractions influence price disparities among neighborhoods.
